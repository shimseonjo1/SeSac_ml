```python
import numpy as np
```

```python
data = np.random.randint(-100,100,50).cumsum()
data
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.rcParams['font.family'] = 'Nanum Pen Script'
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```
